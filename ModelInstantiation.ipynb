```python
from transformers import BertModel, AutoModel, AutoModelForMaskedLM, AutoTokenizer
```

```python
model_name = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
bert_model = BertModel.from_pretrained(model_name)
auto_model = AutoModel.from_pretrained(model_name)
auto_model_masked_lm = AutoModelForMaskedLM.from_pretrained(model_name)
```

```python
sequences = ["Paris is the capital of [MASK]."]
tokens = tokenizer(sequences, padding=True, truncation=True, return_tensors="pt")
```

```python
output1 = bert_model(**tokens)
output2 = auto_model(**tokens)
output3 = auto_model_masked_lm(**tokens)
```

```python
output1 # Last hidden state
```

```python
output2 # Same as output1
```

```python
output3 # Different – this includes logits! MaskedLM means that it includes the classifier at the end.
```
